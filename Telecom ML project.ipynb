# Machine Learning

```python
#Linear Regression
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x = customer_churn[['tenure']]
y = customer_churn[['MonthlyCharges']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,random_state = 0)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
y_pred = pd.DataFrame(y_pred, columns = ['Predicted'])
```

```python
y_pred.head()
```

```python
print(regressor.intercept_) #slope
print(regressor.coef_)  # c
```

```python
from sklearn import metrics
mean_error = metrics.mean_squared_error(y_test, y_pred)
```

```python
mean_error
```

```python
np.sqrt(mean_error)
```

```python
y_pred[:5], y_test[:5]
```

# Logistic Regression

```python
y = customer_churn[['Churn']]
x = customer_churn[['MonthlyCharges']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.35, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(x_train, y_train)
```

```python
y_predict = log_model.predict(x_test)
```

```python
y_pre = pd.DataFrame(y_predict, columns = ['Predicted Value'])
```

```python
y_pre.head()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y_test, y_predict), accuracy_score(y_test, y_predict)
```

# Logistic multi variate

```python
y = customer_churn[['Churn']]
x = customer_churn[['MonthlyCharges','tenure']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(x_train, y_train)
```

```python
y_predict = log_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y_test, y_predict), accuracy_score(y_test, y_predict)
```

# Decision Tree

```python
x = customer_churn[['tenure']]
y = customer_churn[['Churn']]
```

```python
from sklearn.tree import DecisionTreeClassifier
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
my_tree = DecisionTreeClassifier()
my_tree.fit(x_train, y_train)
```

```python
y_predi = my_tree.predict(x_test)
```

```python
y_predi_ = pd.DataFrame(y_predi, columns = ['Predicted'])
y_predi_.head()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```
